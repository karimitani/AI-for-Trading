```python
df = pd.read_csv('time_data/walmart_stock.csv')
```

```python
df.head()
```

```python
# Date columns is not a dt obj
df.info()
```

```python
# We have 'default' format but can pass in other
df['Date'] = pd.to_datetime(df['Date'], format=None)
```

```python
df.info()
```

```python
# This works as well
# df['Date'] = df['Date'].apply(pd.to_datetime)    
```

```python
# Could have done this in read_csv using index_col='Date'
df.set_index('Date', inplace=True)
```

```python
df
```

```python
df.index
```

```python
# Essentially a groupby method, A is year end frequency but there are dozens of options
df.resample(rule='A')
```

```python
df.resample(rule='A').mean()
```

```python
df.resample(rule='A').max()
```

```python
df.resample(rule='A', how=max)
```

```python
def first_day(entry):
    return entry[0]
```

```python
# Could also use lambda
df.resample('A').apply(first_day)
```

```python
df['Close'].resample('A').mean().plot(kind='bar')
```

```python
# Use bar chart because it's not a continuous plot, measures hold aggregate values
df['Close'].resample('M').mean().plot(kind='bar', figsize=(16, 6))
```

```python
df['Open'].plot()
df.rolling(window=7).mean()['Close'].plot(figsize=(16, 6))
```

```python
df['Close'].expanding(7).mean().plot()
```

```python
# Close 20 MA
df['Close: 20 day mean'] = df['Close'].rolling(20).mean()
# Upper = 20MA + 2*std(20)
df['Upper'] = df['Close: 20 day mean'] + 2 * df['Close'].rolling(20).std()
# Lower = 20MA - 2*std(20)
df['Lower'] = df['Close: 20 day mean'] - 2 * df['Close'].rolling(20).std()
```

```python
df[['Close', 'Close: 20 day mean', 'Upper', 'Lower']].tail(200).plot(figsize=(16, 6))
```
